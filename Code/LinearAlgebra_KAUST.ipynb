```python
# y = C + Dt

A = np.array([[1,-1],
              [1,0],
              [1,1] ])
A.shape
b = np.array([-1,-3,1])

#----
x_hat = np.linalg.solve(A.T@A,A.T@b)
print(x_hat)
C = x_hat[0]
D = x_hat[1]
print(C , D )
```

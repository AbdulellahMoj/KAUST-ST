```python
# Reading the data
data = pd.read_csv("/content/Student_Performance.csv")

data.head()
```

```python
# we can see the "Extracurricular Activities" are Yes/No so change it to numbers

data['Extracurricular Activities'] = data['Extracurricular Activities'].astype('category').cat.codes

data.head()
```

```python
# get the data size
data.size
```

```python
# now initialize the X and Y

# .drop means all without this
x = data.drop('Performance Index', axis = 1) # axis for col and 0 for row
y = data['Performance Index']
```

```python
print(type(x))
x.head()
```

```python
y.head()
```

```python
# add the bias in the x data
ones = np.ones(x.shape[0]) # creat the bias col
print(type(x))
x = np.column_stack((ones,x.to_numpy()))
print(type(x))
```

```python
# split using scikit learn

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
# why 0.2 ? so we get 80% 20% every time
# can use 0.3 to get 70 30 ...

```

```python
theta = np.linalg.solve(x_train.T @ x_train,x_train.T @ y_train)
# get the weights "theta"
```

```python
theta
```

```python
# use the prediction formula y = theta0 + theta1 @ n1 +... + thetaN @ xN
# numpy so no need for the long formula just theta @ x_test

prediction = x_test @ theta
```

```python
# get the cost fonction / error
def cost(prediction,y_real):

  error =  y_real - prediction
  squear_error = error**2
  MSE = squear_error.mean()
  return MSE
```

```python
my_cost = cost(prediction, y_test)
print(my_cost)
```

```python
# to see the train error
my_cost_train = cost(x_train@theta,y_train)
my_cost_train
```

```python
from sklearn.linear_model import LinearRegression

```

```python
reg = LinearRegression().fit(x_train,y_train)

print(reg.score(x_train,y_train))
print(reg.score(x_test,y_test))
```

```python
# to see if the mse has any diff
print('our')

print(cost(x_train@theta,y_train))
print(cost(prediction, y_test))
diff = my_cost - my_cost_train
print('fkkf', diff)
print('\nSK linear reg')
print(cost(reg.predict(x_train),y_train))
print(cost(reg.predict(x_test),y_test))
```
